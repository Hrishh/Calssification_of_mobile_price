```python
x=train.iloc[:,:-1]   #taking price_range as Y value as we have to classify over it
y=train.iloc[:,-1]
```

```python
#spliting the data into train and test

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=95)
```

```python
#K
```

```python
classifier=KNeighborsClassifier(n_neighbors=45) #KNN classification to classify mobile over price
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)    
```

```python
# to find the k value
error=[]
#calclating error for k values between 1 and 100
for i in range(3,500,2):#taking odd values for k upto 99
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    pred_y_test=knn.predict(x_test)
    error.append(np.mean(pred_y_test!=y_test))  #mean error will be taken when yTest and ypred doesn't match
```

```python
import matplotlib.pyplot as plt    #always takes k value from those point which have low mean error with highest range
plt.plot(range(3,500,2),error,color='red',linestyle='dashed',marker='o',markerfacecolor='blue',markersize=10)
plt.title('Error rate k value')
plt.xlabel('k value')
plt.ylabel('Mean error')
```
